```python
import numpy as np
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
X = digits.data
y = digits.target
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
def write_answer_1(cross_val_score):
    with open("bagging_answer1.txt", "w") as fout:
        fout.write(str(cross_val_score.mean()))
        
cv_values = cross_val_score(DecisionTreeClassifier(), X, y, cv=10)
print(cv_values.mean())
write_answer_1(cv_values)
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
def write_answer_2(cross_val_score):
    with open("bagging_answer2.txt", "w") as fout:
        fout.write(str(cross_val_score.mean()))
        
bagging_model = BaggingClassifier(DecisionTreeClassifier(), n_estimators=100)
cv_values = cross_val_score(bagging_model, X, y, cv=10)
write_answer_2(cv_values)
```

```python
cv_values.mean()
```

```python
def write_answer_3(cross_val_score):
    with open("bagging_answer3.txt", "w") as fout:
        fout.write(str(cross_val_score.mean()))
bagging_model = BaggingClassifier(DecisionTreeClassifier(), n_estimators=100, max_features=8)
cv_values = cross_val_score(bagging_model, X, y, cv=10)
print(cv_values.mean())
write_answer_3(cv_values)
```

```python
def write_answer_4(cross_val_score):
    with open("bagging_answer4.txt", "w") as fout:
        fout.write(str(cross_val_score.mean()))
        
bagging_model = BaggingClassifier(DecisionTreeClassifier(max_features='sqrt'), n_estimators=100)
cv_values = cross_val_score(bagging_model, X, y, cv=10)
print(cv_values.mean())
write_answer_4(cv_values)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100)
cv_values = cross_val_score(rf_model, X, y, cv=10)
print(cv_values.mean())
```

```python
from matplotlib import pyplot as plt
params = {
    'max_depth': range(1,40,1),
    'max_features': range(5,65,5),
    'n_estimators': range(5,120,5)
}
cv_values = []
for depth in params['max_depth']:
    rf = RandomForestClassifier(max_depth=depth)
    cv_values.append(cross_val_score(rf, X, y, cv=10).mean())
plt.plot(params['max_depth'], cv_values)
plt.title('max_depth')
```

```python
cv_values = []
for n_feat in params['max_features']:
    rf = RandomForestClassifier(max_features=n_feat)
    cv_values.append(cross_val_score(rf, X, y, cv=10).mean())
plt.plot(params['max_features'], cv_values)
plt.title('max_features')
```

```python
cv_values = []
for n_estim in params['n_estimators']:
    rf = RandomForestClassifier(n_estimators=n_estim)
    cv_values.append(cross_val_score(rf, X, y, cv=10).mean())
plt.plot(params['n_estimators'], cv_values)
plt.title('n_estimators')
```
